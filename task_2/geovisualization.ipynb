```python
import pandas as pd
import folium as fol
from geopy.geocoders import Nominatim
```

```python
district = pd.read_csv('data/candidates_district.csv')

len(district)

loc = Nominatim(user_agent="man")

map_center = [6.9271, 79.8612]  # Latitude and longitude of colombo
my_map = fol.Map(location=map_center, zoom_start=8)

for i in range(len(district)):
    getloc = loc.geocode(district['District'][int(i)])
    cout = district['Sat'][int(i)]
    
    marker_location = [getloc.latitude, getloc.longitude]
    fol.Marker(location=marker_location, popup='<h4 style="color:green;">'+district['District'][int(i)]+'</h4><p>count:'+str(cout)+'</p>',tooltip='<strong>Click here to see Popup</strong>').add_to(my_map)
     
my_map
```
